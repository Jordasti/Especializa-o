### 3° Passo - Analisando os termos do texto

```python
#Transformando a lista de palavras em um dicionário e contando os termos que se repetem.
dicionario = {}
for termos in listatexto:
    dicionario[termos] = listatexto.count(termos)
print(dicionario)

print(f'O texto escolhido tem {len(listatexto)} palavras.')
print(f'O texto escolhido tem {len(dicionario)} termos.')
```

### 4° Passo - Geração de gráfico

```python
#Transformando o dicionario criado em um dataframe e ordenando de forma crescente os termos que mais se repetem.
dicionario_df = pd.DataFrame.from_dict(dicionario, orient = 'index')
dicionario_df = dicionario_df.rename(columns={0: 'Quantidade'})
dicionario_df = dicionario_df.sort_values('Quantidade', ascending = False)

#Plotando no grafico os 30 termos que mais se repetem.
dicionario_df[:30].plot(figsize=(15,5), kind='bar')
plt.title('Frequencia de termos')
plt.ylabel('Quantidade de repetições')
plt.xlabel('Termos')
plt.legend()

plt.show()
```
